```python
# OSError
try:
    open("non_existent_file.txt")  # File does not exist
except FileNotFoundError as e:     # More specific
    print("FileNotFoundError caught:", e)
except OSError as e:               # General OS-related error
    print("OSError caught:", e)
```

```python
# NameError
# This example tries to print a variable that was never declared.
try:
    print(var)
except NameError as e:
    print(e)
```

```python
# KeyError
d = {"key1": "value1"}

try:
    val = d["key2"]
except KeyError as e:
    print(e)
```

```python
# IndexError
my_list = [1, 2, 3]

try:
    element = my_list[5]
except IndexError as e:
    print(e)
```

```python
# AttributeError
class MyClass:
    pass

obj = MyClass()

try:
    obj.some_attribute
except AttributeError as e:
    print(e)
```

```python
# AssertionError
try:
    assert 1 == 2, "Assertion failed"
except AssertionError as e:
    print(e)
```

```python
import math
try:
    result = math.exp(1000)  # Exponential function with a large argument
except OverflowError as e:
    print(e)
```

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(e)
```
